```
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action="ignore")
```

```
clusters = pd.read_csv('/content/clustered_data.csv')
data  = pd.read_csv('/content/data-clustering-rackathon.csv')
```

```
SECTORS = ['Shopping' , 'Electronics', 'Food' , 'Pharmacy', 'Insurance' , 'Travel', 'Education', 'Govt. Bills and taxes', 'Misc.']
CLUSTER_DICT = {0: 'who make all type of purchases', 
                1: 'more people with due payments', 
                2: 'who purchases mostly in installments', 
                3: 'who take more cash in advance', 
                4: 'who make expensive purchases',
                5:"who don't spend much money"}
```

```
CLUSTER_TABLE = pd.DataFrame({
    'CUST_ID' : data['CUST_ID'], 
    'CLUSTER' : clusters['cluster']
})

SECTOR_TABLE = pd.DataFrame({
    'CUST_ID' : data['CUST_ID'], 
    'SECTOR' : np.random.choice(SECTORS, size = len(data)), 
})
```

```
print(SECTOR_TABLE.head(), "\n\n")
print(CLUSTER_TABLE.head())
```

```
def GetCustomerBySector(sector_list):
    CUST_LIST = []
    for i in sector_list:
        CUST_LIST += data[SECTOR_TABLE['SECTOR'] == i]['CUST_ID'].tolist()
    return CUST_LIST

def GetCustomersByClusters(cluster_list):
    CUST_LIST = []
    for i in cluster_list:
        CUST_LIST += data[clusters['cluster'] == i]['CUST_ID'].tolist()
    return CUST_LIST
```

```
for i in GetCustomersByClusters([1, 2, 4]):
    print(i, end = '\t')
```

```
for i in GetCustomerBySector(['Shopping']):
    print(i, end = '\t')
```
